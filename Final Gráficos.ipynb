```python
import matplotlib.pyplot as plt

# Dados de exemplo (contagem de vendas por categoria)
categorias = ['Basic', 'Standard', 'Premium']
vendas = [120, 150, 100]

# Cores para as fatias do gráfico de pizza
cores = ['gold', 'lightcoral', 'lightskyblue']

# Explodir a primeira fatia (Basic)
explode = (0.0, 0.1, 0)

# Criar o gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(vendas, labels=categorias, colors=cores, autopct='%1.1f%%', startangle=140, explode=explode)

plt.title('Distribuição de Vendas por Categoria')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV em um DataFrame
df = pd.read_csv('Netflix.csv')

# Criar um DataFrame agregado com contagem de entradas por idade e tipo de assinatura
df_agregado = df.groupby(['Age', 'Subscription Type']).size().unstack(fill_value=0)

# Criar o gráfico de barras empilhadas
ax = df_agregado.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Idade')
plt.ylabel('Contagem')
plt.title('Relação entre Idade e Tipo de Assinatura no Netflix')

plt.show()
```
